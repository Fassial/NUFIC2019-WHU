```python
import pynq
from pprint import pprint

overlay = pynq.Overlay("/home/xilinx/jupyter_notebooks/workshop/m_bit/PYNQ_DMA.bit")
pprint(overlay.ip_dict)
```

```python
from DMA_SG.DMA_SG import DMA_SG

dma_pl2ps_description = overlay.ip_dict['axi_dma_write']
dma_pl2ps = DMA_SG(dma_pl2ps_description)
```

```python
data_size = 400 * 8

buffer = dma_pl2ps.recvchannel.transferS2MM(data_size, restart = 1)
```

```python
from DMA_SG import tools

tools.print_descriptor_attr(dma_pl2ps.recvchannel.descriptor[0])
tools.print_dma_status(dma_pl2ps)
tools.print_buffer(buffer)
buffer = dma_pl2ps.recvchannel.transferS2MM(data_size)
```

```python
from pynq.lib import AxiGPIO

axi_transmit_intr_description = overlay.ip_dict['axi_transmit_intr']
axi_transmit_intr = AxiGPIO(axi_transmit_intr_description)
axi_transmit_intr.setlength(1, channel = 1)
axi_transmit_intr.setdirection(AxiGPIO.Input, channel = 1)
print(axi_transmit_intr.channel1[0])
axi_transmit_intr.channel1[0].wait_for_value(1)
```

```python
print(axi_transmit_intr.channel1[0].read())
print(axi_transmit_intr)
```

```python
import asyncio

@asyncio.coroutine
def update_buffer():
    while True:
        yield from axi_transmit_intr.channel1[0].wait_for_value_async(1)
        tools.print_buffer(buffer)
        buffer = dma_pl2ps.recvchannel.transferS2MM(data_size)
        yield from axi_transmit_intr.channel1[0].wait_for_value_async(0)
```

```python
tasks = [asyncio.ensure_future(update_buffer)]
```
